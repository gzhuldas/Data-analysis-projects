### Task 2

```python
review_rating = '''SELECT 
                    reviews.book_id AS book_id,
                    books.title AS title,
                    count(reviews.review_id) AS review_number,
                    --ratings.book_id AS rating_books_id,
                    AVG(ratings.rating) AS average_rating
                FROM
                    reviews
                INNER JOIN ratings ON reviews.book_id = ratings.book_id
                INNER JOIN books ON books.book_id = ratings.book_id
                --LIMIT 3;
                GROUP BY
                    reviews.book_id,
                    books.title
                ORDER BY
                    average_rating DESC,
                    review_number DESC
'''
review_rating = pd.io.sql.read_sql(review_rating, con = engine)
review_rating
```

The table above shows the total number of reviews as well as the average rating each of the books received. A book named "A Dirty Job (Grim Reaper #1)" has the highest number of reviews and rating. The least popular book is "Harvesting the Heart".

### Task 3

```python
top_publishers = '''SELECT
                     books.publisher_id as publisher_id,
                     publishers.publisher,
                     count(books.book_id) as num_books
                    FROM
                     books
                    INNER JOIN publishers ON publishers.publisher_id = books.publisher_id
                    WHERE
                     books.num_pages > 50
                    GROUP BY
                     books.publisher_id,
                     publishers.publisher
                    ORDER BY
                     num_books DESC
                    LIMIT 3; 
'''
top_publishers = pd.io.sql.read_sql(top_publishers, con = engine)
top_publishers
```

The table above shows top-3 publishers who published the greatest number of books which have more than 50 pages.

### Task 4

```python
top_authors = '''SELECT
                  books.author_id AS author_id,
                  authors.author AS author,
                  rating_number.rating_num AS rating_num,
                  rating_number.average_rating AS average_rating
                  FROM 
                   (SELECT
                     book_id,
                     count(rating_id) as rating_num,
                     AVG(rating) as average_rating
                    FROM
                     ratings
                    GROUP BY
                     book_id
                   ) AS rating_number
                  INNER JOIN books ON rating_number.book_id = books.book_id 
                  INNER JOIN authors ON authors.author_id = books.author_id
                  WHERE 
                   rating_num > 50
                  ORDER BY
                   average_rating DESC
                  LIMIT 3; 

'''
top_authors = pd.io.sql.read_sql(top_authors, con = engine)
top_authors
```

```python
top_authors = '''SELECT
                  authors.author AS author,
                    COUNT(rating_id) AS rating_cnt,
                    AVG(rating) as rating_avg
                  FROM 
                    ratings
                  INNER JOIN books ON ratings.book_id = books.book_id 
                  INNER JOIN authors ON authors.author_id = books.author_id
                    GROUP BY
                    authors.author
                    HAVING 
                    COUNT(rating_id) > 50     
                  ORDER BY
                   rating_avg DESC
                  LIMIT 3; 

'''
top_authors = pd.io.sql.read_sql(top_authors, con = engine)
top_authors
```

The table above shows the authors with the highest number of users who rated the book and the highest average rating. As can be seen J.K. Rowling/Mary GrandPr√© and J.R.R. Tolkien are in the top.

### Task 5

```python
average_text_rating = '''SELECT
                          ratings.username as username,
                          COUNT(ratings.rating_num) as average_num_text
                         FROM
                            (SELECT
                              username,
                              count(rating_id) as rating_num
                            FROM
                              ratings
                            GROUP BY
                             username
                            ) AS ratings
                         INNER JOIN reviews ON ratings.username = reviews.username
                         WHERE 
                          ratings.rating_num > 50
                         GROUP BY
                          ratings.username
'''
average_text_rating = pd.io.sql.read_sql(average_text_rating, con = engine)
average_text_rating
```

The table above shows the number of text reviews each of the users who rated more than 50 books has. 

```python
average_text_rating = '''with reviews as 
                         (SELECT
                           username,
                           COUNT(review_id) as review_num
                         FROM
                           reviews
                         GROUP BY
                           username
                         ),
                         ratings as
                         (SELECT
                           username,
                           COUNT(rating_id) as rating_num
                          FROM
                           ratings 
                          GROUP BY
                           username
                         )
                          SELECT
                           AVG(reviews.review_num) as average_text_review_number
                          FROM
                           reviews
                          INNER JOIN ratings ON ratings.username = reviews.username
                          WHERE
                           ratings.rating_num >= 50

'''
average_text_rating = pd.io.sql.read_sql(average_text_rating, con = engine)
average_text_rating
```
